```python
import torch as tr
from IPython.lib.display import Audio

from experiments.synth import GranularSynth
from synth import generate_chirp_texture
```

```python
duration = 4
# event_duration = 2 ** (-2)
event_duration = 2**1
fmin = 2**8
fmax = 2**11
# n_events = 2**6
n_events = 2**0
sr = 2**13

Q = 12
hop_length = 2**8
seed = 43

theta_density = 1.0
theta_slope = 0.9

texture = generate_chirp_texture(
    theta_density,
    theta_slope,
    duration=duration,
    event_duration=event_duration,
    sr=sr,
    fmin=fmin,
    fmax=fmax,
    n_events=n_events,
    Q=Q,
    hop_length=hop_length,
    seed=seed,
)
Audio(texture, rate=sr)
```

```python
sr = 2 ** 13
duration = 2 ** 2
grain_duration = 2 ** -2
n_grains = 2 ** 6
f0_min_hz = 2 ** 8
f0_max_hz = 2 ** 11
Q = 12
hop_len = 2 ** 8
seed = 42
theta_density = tr.tensor(1.0)
theta_slope = tr.tensor(0.5)

n_samples = int(duration * sr)
grain_n_samples = int(grain_duration * sr)

synth = GranularSynth(sr=sr,
                      n_samples=n_samples,
                      n_grains=n_grains,
                      grain_n_samples=grain_n_samples,
                      f0_min_hz=f0_min_hz,
                      f0_max_hz=f0_max_hz,
                      seed=seed,
                      Q=Q,
                      hop_len=hop_len)

x = synth.forward(theta_density, theta_slope)
print(x.max())
Audio(x, rate=sr)
```
